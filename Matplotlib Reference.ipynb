```python
import numpy as np
import matplotlib.pyplot as plt
```

# Single plot in the figure: *add_subplot()*

Note that the convention here is that we use **ax** instead of **axes** because there is just 1 plot.

```python
fig = plt.figure(figsize=[10,5]) #figsize controls how big the graph appears on the screen
ax = fig.add_subplot() #add_subplot() is for 1 graph on the plot; for multiple graphs on the plot use add_subplots() (with 's')
plt.rcParams.update({'font.size': 20}) #Used to control font size of the values along the axes. This also controls font size of other graph elements (e.g. graph titles, axis titles), which can then be overwritten if needed as shown a few lines below. 

ax.scatter([2,4,6,8,10], [20,40,60,80,100], color='green', marker='*', s=100) #Used for scatterplot
ax.plot([3,5,7], [15,30,60], color='red', linewidth=5) #Used for lineplot
ax.bar([8,9,10],[10,15,20], color='cyan') #Used for bar chart
ax.set_xlim([0,15]) 
ax.set_ylim([0,120])
ax.set_ylabel('My Y-Axis', color='orange', size=15) #Overwrites the font size that was set in plt.rcParams.update({'font.size': 20})
ax.set_xlabel('My X-Axis', color='purple', size=25) #Overwrites the font size that was set in plt.rcParams.update({'font.size': 20})
ax.set_title('Single Plot Example', color='blue', size=40) #Overwrites the font size that was set in plt.rcParams.update({'font.size': 20})
ax.tick_params(size=50) #Controls the size of the axis tick lines
ax.vlines(x=12, ymin=10, ymax=110, linewidth=8, color='black') #Adds a vertical line
ax.hlines(y=110, xmin=1, xmax=3, linewidth=4, color='pink') #Adds a horizontal line
plt.show()
```

# Multiple plots in the figure: *add_subplots()*

-Note the **s** at the end of add_subplot**s**(). </br></br>
-Note the convention here is that we use **axes** instead of **ax** because there are multiple plots. </br>
-Note that we can add plot elements to any of the individual plots just as we did in the single plot example above.

```python
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=[20,10]) #Creates 6 plots, arranged in a 2x3 grid. axes is a 2x3 NumPy array, one for each plot.
plt.rcParams.update({'font.size': 20}) #Used to control font size of the values along the axes. This also controls font size of other graph elements (e.g. graph titles, axis titles), which can then be overwritten if needed as shown a few lines below. 
fig.tight_layout(pad=3) #Controls amount of whitespace between adjacent plots in the grid.

#Any of the 6 plots can be accessed with the proper indexes.

axes[0,0].set_title('Graph 1')
axes[0,1].set_title('Graph 2')
axes[0,2].set_title('Graph 3')
axes[1,0].set_title('Graph 4')
axes[1,1].set_title('Graph 5')
axes[1,2].set_title('Graph 6')

#We can add graphs elements to any of the 6 plots similarly as in the single plot example above.

axes[0,1].scatter([2,4,6,8,10], [20,40,60,80,100], color='green', marker='*', s=100) #Used for scatterplot
axes[0,1].plot([3,5,7], [15,30,60], color='red', linewidth=5) #Used for lineplot
axes[0,1].bar([8,9,10],[10,15,20], color='cyan') #Used for bar chart
axes[0,1].set_xlim([0,15]) 
axes[0,1].set_ylim([0,120])
axes[0,1].set_ylabel('My Y-Axis', color='orange', size=15) #Overwrites the font size that was set in plt.rcParams.update({'font.size': 20})
axes[0,1].set_xlabel('My X-Axis', color='purple', size=15) #Overwrites the font size that was set in plt.rcParams.update({'font.size': 20})
axes[0,1].set_title('Single Plot Example', color='blue', size=15) #Overwrites the font size that was set in plt.rcParams.update({'font.size': 20})
axes[0,1].tick_params(size=8) #Controls the size of the axis tick lines
axes[0,1].vlines(x=12, ymin=10, ymax=110, linewidth=8, color='black') #Adds a vertical line
axes[0,1].hlines(y=110, xmin=1, xmax=3, linewidth=4, color='pink') #Adds a horizontal line

plt.show()
```
